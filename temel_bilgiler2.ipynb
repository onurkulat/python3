DECORATOR FONKSİYONLARIN OLUŞTURULMASI VE KULLANILMASI

```python
import time
def kareleri_hesapla(sayılar):
    sonuç=list()
    baslama=time.time()
    for i in sayılar:
        sonuç.append(i**2)
    bitis=time.time()
    print("Bu fonksiyon "+str(bitis-baslama)+" saniye sürdü")
    return sonuç

def küpleri_hesapla(sayılar):
    sonuç=list()
    baslama=time.time()
    for i in sayılar:
        sonuç.append(i**3)
    bitis=time.time()
    print("Bu fonksiyon "+str(bitis-baslama)+" saniye sürdü")
    return sonuç
```

```python
kareleri_hesapla(range(10))
küpleri_hesapla(range(10))
```

```python
#decorator ile ;
```

```python
import time

def zaman_hesapla(func):
    def wrapper (sayılar):
        baslama=time.time()
        sonuç=func(sayılar)
        bitis=time.time()
        print(func.__name__+" "+str(bitis-baslama)+" saniye sürdü.")
        return sonuç
    return wrapper

@zaman_hesapla
def kareleri_hesapla(sayılar):
    sonuç=list()
    for i in sayılar:
        sonuç.append(i**2)
    return sonuç

@zaman_hesapla
def küpleri_hesapla(sayılar):
    sonuç=list()
    for i in sayılar:
        sonuç.append(i**3)
    return sonuç
```

```python
kareleri_hesapla(range(10))
küpleri_hesapla(range(10))
```

```python
#örnek
class Kumanda():
    def __init__(self,kanal_listesi):
        self.kanal_listesi=kanal_listesi
        self.index=-1
    def __iter__(self):
        return self
    def __next__(self):
        self.index+=1
        if self.index<len(self.kanal_listesi):
            return self.kanal_listesi[self.index]
        else:
            self.index=-1
            raise StopIteration
```

```python
import requests
from bs4 import BeautifulSoup
url="https://www.imdb.com/chart/top"
response=requests.get(url)
html_icerigi=response.content
soup=BeautifulSoup(html_icerigi,"html.parser")

basliklar=soup.find_all("td",{"class":"titleColumn"})

ratingler=soup.find_all("td",{"class":"ratingColumn imdbRating"})

for baslik,rating in zip(basliklar,ratingler):
    baslik=baslik.text
    baslik=baslik.strip()
    baslik=baslik.replace("\n","")
    rating=rating.text
    rating=rating.strip()
    rating=rating.replace("\n","")
    
    print("Başlık:",baslik)
    print("Rating:",rating)
```

SMTP MODÜLÜ İLE MAİL GÖNDERME

```python
import smtplib
from email.mime.multipart import MIMEMultipart#mail yapısı için
from email.mime.text import MIMEText#mesaj gövdesi
import sys

mesaj=MIMEMultipart()#Mail yapımızı oluşturuyoruz
mesaj["From"]="onrkulat@gmail.com"
mesaj["To"]="onkulat@gmail.com"
mesaj["Subject"]="Smtp Mail Gönderme"

yazi="""
Smtp ile mail gönderiyorum
Onur Kulat
"""

mesaj_govdesi=MIMEText(yazi,"plain")
mesaj.attach(mesaj_govdesi)

try:
    mail=smtplib.SMTP("smtp.gmail.com",587)
    mail.ehlo()#bağlanıyoruz
    mail.starttls()#mailin şifrelenmesi için
    mail.login("onrkulat","şifre")
    mail.sendmail(mesaj["From"],mesaj["To"],mesaj.as_string())
    print("Mail gönderildi..")
    mail.close()
except:
    sys.stderr.write("Bir sorun oluştu!")
    sys.stderr.flush()
    
```
